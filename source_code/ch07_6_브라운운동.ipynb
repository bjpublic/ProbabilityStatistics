```python
import numpy as np

np.random.seed(777)

n = 1000
Bt = np.random.normal(loc=0, scale=1, size=n-1)
Bt = np.insert(Bt, obj=0, values=0)
Bt_cumulative_sum = Bt.cumsum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))

plt.subplot(2, 1, 1)
plt.plot(Bt)
plt.title("Brownian Motion Increments")

plt.subplot(2, 1, 2)
plt.plot(Bt_cumulative_sum)
plt.title("Brownian Motion Path")
plt.show()
```

```python
mean_Bt = np.mean(Bt)
var_Bt = np.var(Bt)
print(mean_Bt)
print(var_Bt)
```

```python
n = len(Bt)
t = 2
B2 = []
for i in range(t,n-1):
    B = np.sum(Bt[i-t+1:i+1])
    B2.append(B)
```

```python
print(np.mean(B2))
print(np.var(B2))
```
