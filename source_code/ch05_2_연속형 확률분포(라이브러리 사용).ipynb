# 연속형 균일 분포

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
```

```python
a = 1
b = 5

x_list = np.arange(a, b, 0.01)
p_list = []

for x in x_list:
    prob = uniform.pdf(x, loc=a, scale=b-a)
    p_list.append(prob)
```

```python
plt.plot(x_list, p_list)
plt.show()
```

## 연속형 균일 분포 난수 생성

```python
import numpy as np
import matplotlib.pyplot as plt

x_list = np.random.uniform(low=1.0, high=5.0, size=10000)
print(x_list)
```

```python
plt.hist(x_list)
plt.show()
```

# 정규 분포

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
mu = 75
s = 3

x_list = np.arange(mu-3*s, mu+3*s, 0.01)
p_list = []


for x in x_list:
    prob = norm.pdf(x, loc=mu, scale=s)
    p_list.append(prob)
```

```python
plt.plot(x_list, p_list)
plt.show()
```

# 감마 분포

```python
import numpy as np
from scipy.stats import gamma
import matplotlib.pyplot as plt
```

```python
alpha = 2
beta = 1

x_list = np.arange(0, alpha*beta*5, 0.01)
p_list = []


for x in x_list:
    prob = gamma.pdf(x, a=alpha, scale=beta)
    p_list.append(prob)
```

```python
plt.plot(x_list, p_list)
plt.show()
```

# 카이제곱 분포

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
```

```python
p = 3

x_list = np.arange(0, p*5, 0.01)
p_list = []


for x in x_list:
    prob = chi2.pdf(x, df=p)
    p_list.append(prob)
```

```python
plt.plot(x_list, p_list)
plt.show()
```

# 베타 분포

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
a = 0.5
b = 0.5

xs = np.arange(0.01, 1, 0.01)
ps = []


for x in xs:
    prob = beta.pdf(x, a=a, b=b)
    ps.append(prob)
```

```python
plt.plot(xs, ps)
plt.show()
```
